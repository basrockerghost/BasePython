```python
import gensim
```

```python
import glob
file_names = glob.glob("search/*.txt")
file_names
```

```python
raw_documents=[]
for file in file_names:
    try:
        with open (file, "r" , encoding = "utf-8") as f: raw_documents.append(f.read())
    except:
        pass
```

```python
print("Number of documents:",len(raw_documents))
```

```python
clean_text=[]
for text in raw_documents :
    clean_text.append(gensim.utils.simple_preprocess(text))
```

```python
dictionary = gensim.corpora.Dictionary(clean_text)
print ("Number of  words in dictionary",len(dictionary))

for i in range (100):
    print (i , dictionary[i])
    
```

```python
print(dictionary)
```

```python
corpus = [dictionary.doc2bow(text) for text in clean_text]
print (corpus[:10])
```

```python
tf_idf=gensim.models.TfidfModel(corpus)
print(tf_idf)
```

```python
similarity_object = gensim.similarities.Similarity('search/',tf_idf[corpus],num_features=len(dictionary))
print(similarity_object)
print(type(similarity_object))
```

```python
text = raw_documents[1]
query_doc = gensim.utils.simple_preprocess('percy')
print(query_doc)
```

```python
query_doc_bow =  dictionary.doc2bow(query_doc)
print(query_doc_bow)
```

```python
query_doc_tf_idf=tf_idf[query_doc_bow]
print(query_doc_tf_idf)
```

```python
similarty_scores=list(similarity_object[query_doc_tf_idf])
similarty_scores
```

```python
max_scores=max(similarty_scores)
print(max_scores)
similarty_scores.index(max_scores)
```

```python
print ('{:.150}'.format(raw_documents[similarty_scores.index(max_scores)]),"...Show more")
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2,
    workers=4,
)
```

```python
model.build_vocab(clean_text, progress_per=1000)
```

```python
model.train(clean_text, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.save("./word2vec-amazon-cell-accessories-reviews-short.model")
```
