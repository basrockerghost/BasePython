```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs

n_samples = 300
X, y = make_blobs(n_samples = n_samples, centers = 2, random_state=42, cluster_std=1.0)
anomaly = np.array([[5, 5], [6, 6], [7, 7]])

print(f'Result of X : {X[:10]}')
print(f'Result of y : {y[:100]}')
```

```python
X = np.vstack([X, anomaly])
```

```python
plt.scatter(X[:, 0], X[:, 1], c = 'b', marker = 'o', s = 25)
plt.title("Visualize the Normal and Anomaly data")
plt.show()
```

```python
#use dbscan for find anomaly
dbscan = DBSCAN(eps = 1, min_samples= 40)
labels = dbscan.fit_predict(X)

#anomaly show -1
anomaly = X[labels == -1]
```

```python
#show anomaly
plt.scatter(X[:, 0], X[:, 1], c = 'b', marker= 'o', s = 25)
plt.scatter(anomaly[:, 0], anomaly[:, 1], c = 'r', marker= 'x', s = 50, label= 'Anomalies')
plt.title('Anomaly Detection with DBScan')
plt.legend()
plt.show()
```
